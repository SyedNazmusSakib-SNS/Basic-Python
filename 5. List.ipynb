## Nesting Lists

Nesting Lists is one of the great features in Python data structures. Nesting Lists means we can have data structures within data structures. 


```python
# Let's make three lists
lst_1=[1,2,3]
lst_2=[4,5,6]
lst_3=[7,8,9]

# Make a list of lists to form a matrix
matrix = [lst_1,lst_2,lst_3]
```

```python
# Show
matrix
```

```python
matrix[2][1]
```

We can re-use indexing to grab elements, but now there are two levels for the index. 

**a.** The items in the matrix object

**b.** The items inside the list

```python
# Grab first item in matrix object
matrix[0]
```

```python
# Grab first item of the first item in the matrix object
matrix[0][0]
```

Many times people find the difference between extend and append to be unclear. So note that,

**append: Appends object at end**

```python
x = [1, 2, 3]
x.append([4, 5])
print(x)
```

**extend: extends list by appending elements from the iterable. It will append content of list elements to the existing list**

```python
x = [1, 2, 3]
x.extend('ty')
print(x)
```

```python
l1 = ["ineuron","DS","AI"]
l2 = list(range(5))
print("extended list")
l1.extend(l2)
l1
```

### Iterating through a string by using list Comprehension

```python
# Without using list comprehension
letter = []
for ch in 'Sakib':
    letter.append(ch)
letter
```

```python
# With using list comprehension
letter = [ ch for ch in 'Sakib' ]
print(letter)
```

Use of conditional statements using list comprehension

```python
#Without list comprehension
l=[]
for i in range(20):
    if i%2 == 0:
        l.append(i)
l
```

```python
# With list comprehension
number = [ i for i in range(20) if i % 2 == 0]
print(number)
```

```python
l = []
for i in range(20):
    if i%2 != 0:
        l.append("Odd")
    else:
        l.append("Even")
l
```

```python
# if else in list comprehension
odd_even = ["Odd" if i%2!=0 else "Even" for i in range(20)]
odd_even
```

```python
# Nested for loop without list comprehention
matrix = []
for i in range(3):
    matrix.append([])
    for j in range(3):
        matrix[i].append(j)

matrix
```

```python
# Nested for loop with list comprehention
matrix_with_List_Comp = [[j for j in range(3)] for i in range(3)]
matrix_with_List_Comp
```

```python
# Nested conditions
[i for i in range(8) if i%2==0 if i%3==0]
```

```python
l = []
for i in range(8):
    if i%2 ==0:
        if i%3 == 0:
            l.append(i)
l
```

```python
# For if elif else
l = [1, 2, 3, 4, 5, 6, 7, 8, 9]
['yes' if i == 1 else 'no' if i == 2 else 'idle' for i in l]
```

```python
l = [1, 2, 3, 4, 5, 6, 7, 8, 9]
res = []
for i in l:
    if i == 1:
        res.append("yes")
    elif i == 2:
        res.append("no")
    else:
        res.append("idle")
        
res
```
